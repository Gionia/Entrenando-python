```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=range(10)
```

```python
x=np.arange(2,50,2)
plt.plot(x,x,"r--",x,2*x,"g-")
plt.axis([0,50,0,50])
plt.show()
```

```python
f=lambda x: np.exp(-x)*np.cos(2*np.pi*x)
```

```python
vector1=np.arange(0,5,.1)
vector2=np.arange(0,5,.2)
```

```python
plt.figure(2)
plt.subplot(211)
plt.plot(vector1,f(vector1),'ro',vector2,f(vector2),'k')
plt.figure(1)
plt.subplot()
plt.plot([1,2,3])
plt.show()
```
