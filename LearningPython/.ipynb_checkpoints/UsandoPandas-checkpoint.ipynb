# Manipulando los datos perdidos con python

## Formas de datos perdidos en python
*Los datos perdidos pueen aparecer de las siguientes maneras:*
* `?`
* `N/A`
* `0`
* ` `
* `Nan`

*Lo primero que debes hacer es verificar con la persona o grupo que recopiló los datos si es posible que se pueda encontrar el valor real que está faltante.*

*Otra posibilidad es simplemente eliminar los datos donde se encuentra ese valor perdido. Acá tienes dos opciones, la primera sería eliminar la variable faltante o eliminar la fila completa en donde está se encuentre. Esta opción es conveniente en los casos que no se tiene muchos datos faltantes. Toma en cuenta que se debe buscar hacer el menor impacto posible, por lo que se debe tener cuido al momento de eliminar datos.*

## Eliminando datos faltantes

````python
df2.dropna(axis=0) #Elimina las filas con valores vacios
````

````python
df2.dropna(axis=1) #elimina las columnas con valores perdidos
````

````python
df2.dropna(subset=['Cabina'],axis=0) 
#Revisa si existen valores vacion en la columna 'cabina'
#y si es así elimina las filas correspondientes
````

**Si utilizamos el metodo de eliminación es importnte tener en cuenta que los indices de las filas borradas no mueven a los emas indices es decir puede quedar 1 4 5 6 7 15**
*Para que esto no suceda implementamos*
`df2.dropna(axis=0,inplace=True)`

# El agrupamiento de datos o binning 

*El agrupamiento de datos o binning en ingles, es un método de preprocesamiento de datos y consiste en agrupar valores en compartimientos. En ocasiones este agrupamiento puede mejorar la precisión de los modelos predictivos y, a su vez, puede mejorar la comprensión de la distribución de los datos.*

En pandas el comando para agrupar datos es:

`pd.cut()`

````python
df3=pd.read_csv('/home/giovannyencinia/Documents/DataSets/titanic/train.csv')
df3.columns=('ID','Sobreviviente','Clase','Nombre','Sexo','Edad','Hermanos','Hijos','Ticket','Tarifa','Cabina','Embarque')
````

````python
bins = [0, 5, 12, 18, 35, 60, 100]#el primer numero es el número con el que se inicia el rango, despues los ultimos numero de los intervalos
names=["1","2","3","4","5","6",'7']
df3['Edad']=pd.cut(df3['Edad'],bins,labels=names)#agrupamos en intervalos
df3
````
