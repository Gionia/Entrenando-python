```python
from matplotlib import pyplot as plt
from ipywidgets import interact
%matplotlib inline
```

```python
peso = [1, 2, 3, 4]
altura = [1, 4, 9, 16]
altura1 = [2 * i for i in altura]
plt.plot(peso, altura, 'b-.', label = "mayores a 80 anios")
plt.plot(peso, altura1, label = "muerto")
plt.xlabel("Temperatura")
plt.ylabel("Pelicula con 8 estrellas")
plt.legend()
plt.xlim(-1, 30)
plt.ylim(-1, 30)
plt.savefig("Primer-grafica.pdf")
plt.show()
```

```python
def barras(wth = .5):
    plt.bar(peso, altura, width = wth)
    plt.show()
```

```python
interact(barras)
None
```
