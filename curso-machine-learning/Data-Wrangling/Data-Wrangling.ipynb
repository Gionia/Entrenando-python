# Dummie DataFrame con variables categóricas

```python
gender = ["Male", "Female"]
income = ["Poor", "Medium class", "Rich"]
number = 500
gender_data = np.random.choice(gender, number)
income_data = np.random.choice(income, number)
weight_data = 160.0 + 30.0 * np.random.randn(number)
height_data = 65.0 + 25.0 * np.random.randn(number)
age_data = 30 + 12 * np.random.randn(number)
df_cate_dummie = pd.DataFrame({'Gender': gender_data, 'Economic Status': income_data,
                               'Weight': weight_data, 'Height': height_data,
                               'Age': age_data})
```

```python
df_cate_dummie.head()
```

```python
double_group = df_cate_dummie.groupby(['Gender', 'Economic Status'])
double_group.aggregate([np.sum])
```
