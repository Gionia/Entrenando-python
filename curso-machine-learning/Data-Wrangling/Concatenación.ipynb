# Merge

```python
path = root + '/athletes'
df_main = pd.read_csv(path + '/Medals.csv', encoding = 'ISO-8859-1')
df_country = pd.read_csv(path + '/Athelete_Country_Map.csv', encoding='ISO-8859-1')
df_sport = pd.read_csv(path+ '/Athelete_Sports_Map.csv', encoding='ISO-8859-1')
```

```python
df_country_dp = df_country.drop_duplicates('Athlete')
df_sport_dp = df_sport.drop_duplicates('Athlete')
df_main_country = pd.merge(left = df_main, right = df_country_dp,
                           on='Athlete')
df_olimpics = pd.merge(left=df_main_country, right=df_sport_dp, 
                       on='Athlete')
```
